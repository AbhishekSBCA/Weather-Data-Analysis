```python
#what is the means temp recorder by month
#one solution
mean_temperature={}

for month in range(1,13):
    mean_temperature[month]=df.loc[df.index.month==month,'Temp_C'].mean()
pd.Series(mean_temperature)
```

```python
#or using pivot 
mean_temperature_df=df.pivot_table(values='Temp_C', index=df.index.month,aggfunc='mean')
mean_temperature_df
```

```python
#or using groupby
mean_temperature_df=df.groupby(df.index.month).mean(numeric_only=True).reset_index()
mean_temperature_df['Temp_C']
```

```python
#extract month name from date and store it in new column named as month
df['Month']=df.index.month
df['Month']=df['Month'].apply(lambda x:calendar.month_name[x])
df.head()
```

```python
#get the list of the months in whichh it rained with thunderstorm
def checklist(x):
    weather_list=['rain','thunderstorms']
    count=0
    x = str(x).lower()
    for substring in weather_list:
        if substring in x:
            count += 1
    if(count==2):
        return True
    else:
        return False
    
print ("2) get the list of the months in whichh it rained with thunderstorm\n" )
rained=df['Weather'].apply(checklist)
month_list=df[rained]['Month'].unique()
print(month_list)
```
